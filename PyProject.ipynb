```python
class Etudiant:
    """Cette classe permet d'enregistrer un étudiant et de l'immatriculer"""
    _baseDeDonnees = list()
    def __init__(self,nom,prenom,adresse,dateNaissance):
        """Le constructeur initialise les champs de la classe"""
        self._matricule = ""
        self._nom = nom
        self._prenom = prenom
        self._adresse = adresse
        self._dateNaissance = dateNaissance
        self._initializelist()
        self._premierChamp()
        self._deuxiemeChamp()
        self._troisiemeChamp()
        self._formerMatricule()
        
    def _initializelist(self):
        """Cette méthode recharge les matricules existants dans une liste à partir d'un fichier texte"""
        if(len(self._baseDeDonnees) == 0):
            fichier = open("EsAnMiSa_2022.txt",'r')
            ch = fichier.read()
            ch1 = ch.split('\n')
            l = len(ch1)
            for i in range(2, l-1):
                ch2 = ch1[i].split('::')
                self._baseDeDonnees.append(ch2[0])
            fichier.close()
            
    def _premierChamp(self):
        """Cette méthode forme le premier champ du matricule à partir des trois premières consonnes du nom.
        En cas de lettres manquantes, elle utilise le caractère 'X'"""
        ch = self._nom.upper()
        voyelles = ['A', 'E', 'I', 'O', 'U', 'Y']
        chaine = ''
        i = 0
        j = 0
        while(i < len(ch) and j < 3):
            if(ch[i] not in voyelles):
                chaine += ch[i]
                j += 1
            i += 1
        for i in range(0, 3-j):
            chaine += 'X'
        self._matricule += chaine
    
    def _deuxiemeChamp(self):
        """Cette méthode forme le deuxième champ du matricule à partir des trois premières lettres du prénom.
        En cas de lettres manquantes, elle utilise le caractère 'X'"""
        ch = self._prenom.upper()
        count = 0
        resultat = ""
        for x in ch:
            if(count == 3):
                break
            resultat = resultat + x
            count += 1
        if(count < 3):
            while(count < 3):
                resultat = resultat + 'X'
                count += 1
        self._matricule = self._matricule + resultat
        
    def _troisiemeChamp(self):
        """Cette méthode forme le troisième champ du matricule en tenant compte des contraintes du métier"""
        champ_un_deux = self._matricule[:3] + self._matricule[3:6]
        found = False
        for i in range(len(self._baseDeDonnees)-1, -1, -1):
            if(champ_un_deux in self._baseDeDonnees[i]):
                nombre = (self._baseDeDonnees[i].split(champ_un_deux))[1]
                tmp = int(nombre) + 1
                found = True
                if(tmp < 10):
                    self._matricule += ('00' + str(tmp))
                elif(tmp < 100):
                    self._matricule += ('0' + str(tmp))
                else:
                    self._matricule += str(tmp)
                break
        if(found == False):
            self._matricule += '001'
        
    def _formerMatricule(self):
        """Cette méthode enregistre le matricule de l'étudiant dans la liste et enregistre toutes ses informations
        dans le fichier"""
        fichier = open("EsAnMiSa_2022.txt",'a')
        fichier.close()
        fichier = open("EsAnMiSa_2022.txt",'r')
        ch = fichier.read()
        fichier.close()
        fichier = open("EsAnMiSa_2022.txt",'a')
        if (len(ch) == 0):
            fichier.write("MATRICULE::NOM::PRENOM::ADRESSE::DATE DE NAISSANCE\n\n")
        fichier.write(self._matricule + "::" + self._nom + "::" + self._prenom + "::" + self._adresse + "::" + self._dateNaissance + "\n")
        fichier.close()
        self.__class__._baseDeDonnees.append(self._matricule)
    
    def getMatricule(self):
        """Retourne le matricule de l'objet courant"""
        return self._matricule
```
